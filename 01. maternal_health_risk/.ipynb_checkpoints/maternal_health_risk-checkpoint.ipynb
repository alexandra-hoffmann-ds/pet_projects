```python
def get_info(data):
    
    """
    Функция get_info выводит на экран пять случайных строк таблицы и общую информацию, полученную с помощью info()

    Аргументы:
    data (DataFrame): датафрейм

    """
    
    print('Пять случайных строк таблицы:')
    display(data.sample(5))
    print(' ')
    print('Общая информация о таблице:')
    print(' ')
    print(data.info())
```

```python
def preprocess(data, data_name):
    
    """
    Функция preprocess выводит на экран количество дубликатов и пропусков в датафрейме

    Аргументы:
    data (DataFrame) : датафрейм
    data_name (str) : название датафрейма
    
    """
    
    for col in data:
        isna_count = data[col].isna().sum()
        if isna_count > 0:
            print(f'Пропусков найдено в столбце {col}: {isna_count}')
    isna_total = data.isna().sum().sum()
    if isna_total == 0:
        print(f'В таблице {data_name} не найдено ни одного пропуска!')
        
    duplicates_count = data.duplicated().sum()
    if duplicates_count > 0:
        print(f'Дубликатов найдено в таблице {data_name}: {duplicates_count}')
    else:
        print(f'В таблице {data_name} не найдено ни одного дубликата!')
```

```python
def unique(data, data_name, col):
    
    """
    Функция unique выводит на экран список уникальных значений в столбце датафрейма

    Аргументы:
    data (DataFrame) : датафрейм
    data_name (str) : название датафрейма
    col (str) : название столбца
     
    """
    
    print(f"Уникальные значения в столбце '{col}' таблицы {data_name}:\n{data[col].unique()}")
```

```python
def cols_list(data, data_name):
    
    """
    Функция cols_list выводит на экран список столбцов датафрейма
    
    Аргументы:
    data (DataFrame) : датафрейм
    data_name (str) : название датафрейма
    
    """

    print(f'Cписок столбцов в таблице {data_name}:\n\n', data.columns.to_list())
```

```python
def describe(data, col, data_name):
                    
    """
    Функция describe выводит на экран числовое описание данных в количественном столбце

    Аргументы:
    data (DataFrame) : датафрейм
    col (str) : столбец датафрейма
    data_name (str) : название датафрейма
    
    """
    
    print(f'Числовое описание столбца {col} таблицы {data_name}:')
    print(' ')
    print(data[col].describe())
```

```python
def value_counts(data, col, data_name):
                
    """
    Функция value_counts выводит на экран числовое распределение данных в категориальном столбце

    Аргументы:
    data (DataFrame) : датафрейм
    col (str) : столбец датафрейма
    data_name (str) : название датафрейма
    
    """
    
    print(f'Распределение данных в столбце {col} таблицы {data_name}:')
    print(' ')
    print(data[col].value_counts())
```

```python
def hist_violin(data, col, bins, xlabel, title, figsize=(16, 4)):
                        
    """
    Функция hist_violin выводит на экран гистограмму и скрипичную диаграмму

    Аргументы:
    data (DataFrame) : датафрейм
    col (str) : столбец датафрейма
    bins (int) : количество корзин для гистограммы
    xlabel (str) : подпись оси X
    title (str) : заголовок
    figsize : размер графика (по умолчанию 16x4)
    
    """
    
    sns.set(style='darkgrid', palette='pastel')
    fig, ax = plt.subplots(1, 2, figsize=figsize)

    sns.histplot(data=data, x=col, kde=True, bins=bins, color='#abc9eb')
    plt.title('Гистограмма', fontsize=14)
    plt.ylabel('Количество значений', fontsize=11)
    plt.xlabel(xlabel, fontsize=11)
    ax[0].grid()
    
    sns.violinplot(x=data[col], ax=ax[0], color='#abc9eb')
    ax[0].set_title('Скрипичная диаграмма', fontsize=14)
    ax[0].set_ylabel('Количество значений', fontsize=11)
    ax[0].set_xlabel(xlabel, fontsize=11)
    ax[0].grid()
    
    ax_title = fig.add_axes([0.5, 1.0, 0.1, 0.1])
    ax_title.set_axis_off()
    ax_title.text(0, 0, title, ha='center', va='center', fontsize=18)

    plt.show()
```

```python
def barplot(data, col, xlabel, ylabel, title, figsize=(8, 4.5), normalize=False):
    
    """
    Функция barplot выводит на экран столбчатую диаграмму
    
    Аргументы:
    data (DataFrame) : датафрейм
    col (str) : столбец датафрейма
    xlabel (str) : подпись оси Х
    ylabel (str) : подпись оси Y
    title (str) : заголовок
    figsize : размер графика (по умолчанию 8x4.5)
    normalize (bool) : True/False (по умолчанию False)
    
    """
    
    sns.set(style='darkgrid', palette='pastel')
    plt.figure(figsize=figsize)
    
    data_new = data[col].value_counts(normalize=normalize)
    data_new_ = data_new.reset_index()
    data_new_.columns = [col, 'count']
    
    ax = sns.barplot(x=col, y='count', data=data_new_, color='#abc9eb')
    for p in ax.patches:
        if normalize:
            ax.text(p.get_x() + p.get_width() / 2., p.get_height(), 
                    f'{int(p.get_height() * 100)}%', 
                    ha='center', va='bottom', fontsize=8)
        else:
            ax.text(p.get_x() + p.get_width() / 2., p.get_height(), 
                    f'{int(p.get_height())}', 
                    ha='center', va='bottom', fontsize=8)
    
    ax.set_yticklabels([])
    ax.set_title(title, fontsize=16)
    ax.set_xlabel(xlabel, fontsize=11)
    ax.set_ylabel(ylabel, fontsize=11)
    
    plt.tight_layout()
    plt.show()
```

Для упрощения восприятия результатов выполним преобразование признака `body_temp`. Для этого напишем удобную функцию, конвертирующую градусы Фаренгейта в градусы Цельсия и наоборот:

```python
def convert_temp(data, col, conversion_type):
    
    """
    Функция convert_temp переводит температуру из градусов Фаренгейта (℉) в градусы Цельсия (°C) и наоборот

    Аргументы:
    data (DataFrame) : датафрейм
    col (str) : название столбца, содержащего температурные значения
    conversion_type (str) : тип преобразования ('FtoC'/'CtoF')

    """

    if conversion_type == 'FtoC':
        data[col] = (data[col] - 32) * 5 / 9
    elif conversion_type == 'CtoF':
        data[col] = (data[col] * 9 / 5) + 32
    return data
```

```python
# выполним преобразование признака body_temp и округлим полученные значения до одного знака после запятой

risk = convert_temp(risk, 'body_temp', 'FtoC').round(1)
```

```python
def train_test_size(X_train, X_test, y_train, y_test):
                                    
    """
    Функция train_test_size выводит на экран размерность разделённых на выборки данных
    
    Аргументы:
    X_train (DataFrame) : входные признаки тренировочной выборки
    X_test (DataFrame) : входные признаки тестовой выборки
    y_train (Series) : целевой признак тренировочной выборки
    y_test (Series) : целевой признак тестовой выборки
    
    """

    print('Тренировочная выборка')
    print(f'Размерность матрицы со входными признаками: {X_train.shape}')
    print(f'Размерность матрицы с целевым признаком: {y_train.shape}')
    print(' ')
    print('Тестовая выборка')
    print(f'Размерность матрицы со входными признаками: {X_test.shape}')
    print(f'Размерность матрицы с целевым признаком: {y_test.shape}')
```

---

```python
# разделим данные на входные и целевой признаки

X = risk.drop('risk_level', axis=1)
y = risk['risk_level']
```

```python
# обозначим в переменную num_cols те признаки, которые необходимо закодировать

# cols_to_scale = ['age', 'systolic_bp', 'diastolic_bp', 'bs', 'body_temp', 'heart_rate', 'pulse_pressure']
cols_to_scale = ['age', 'systolic_bp', 'diastolic_bp', 'bs', 'body_temp', 'heart_rate']
```

```python
# закодируем признаки из списка num_cols с помощью MinMaxScaler

scaler = MinMaxScaler()
X[cols_to_scale] = scaler.fit_transform(X[cols_to_scale])
```

```python
# разделим данные на тренировочную и тестовую выборки

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)
```

```python
def model_results(model_name, y_train, y_train_pred, y_test, y_test_pred):
    
    """
    
    Функция model_results вычисляет и выводит метрики классификации (Accuracy, Precision, Recall, F1-score),
    выводит полный отчет classification_report для тренировочной и тестовой выборок
    
    Аргументы:
    model_name (str) : название модели
    y_train (array-like) : истинные значения целевой переменной на тренировочной выборке
    y_train_pred (array-like) : предсказанные значения целевой переменной на тренировочной выборке
    y_test (array-like) : истинные значения целевой переменной на тестовой выборке
    y_test_pred (array-like) : предсказанные значения целевой переменной на тестовой выборке
    
    """
    
    accuracy_train = accuracy_score(y_train, y_train_pred).round(4)
    precision_train = precision_score(y_train, y_train_pred, average='weighted').round(4)
    recall_train = recall_score(y_train, y_train_pred, average='weighted').round(4)
    f1_train = f1_score(y_train, y_train_pred, average='weighted').round(4)
    report_train = classification_report(y_train, y_train_pred, zero_division=0, output_dict=True)
    
    print(f'Метрики модели {model_name} на тренировочной выборке:')
    print('')
    print(f'Accuracy: {accuracy_train}')
    print(f'Precision: {precision_train}')
    print(f'Recall: {recall_train}')
    print(f'F1: {f1_train}')
    print('')
    for class_label, metrics in report_train.items():
        if class_label not in ['accuracy', 'macro avg', 'weighted avg']:
            print('-' * 30)
            print(f"Метрики класса {class_label}:")
            print(f"  Precision: {metrics['precision']:.4f}")
            print(f"  Recall:    {metrics['recall']:.4f}")
            print(f"  F1-score:  {metrics['f1-score']:.4f}")
            print('-' * 30)
    print('')
    print('')
    
    accuracy_test = accuracy_score(y_test, y_test_pred).round(4)
    precision_test = precision_score(y_test, y_test_pred, average='weighted').round(4)
    recall_test = recall_score(y_test, y_test_pred, average='weighted').round(4)
    f1_test = f1_score(y_test, y_test_pred, average='weighted').round(4)
    report_test = classification_report(y_test, y_test_pred, zero_division=0, output_dict=True)
    
    print(f'Метрики модели {model_name} на тестовой выборке:')
    print('')
    print(f'Accuracy: {accuracy_test}')
    print(f'Precision: {precision_test}')
    print(f'Recall: {recall_test}')
    print(f'F1: {f1_test}')
    print('')
    for class_label, metrics in report_test.items():
        if class_label not in ['accuracy', 'macro avg', 'weighted avg']:
            print('-' * 30)
            print(f"Метрики класса {class_label}:")
            print(f"  Precision: {metrics['precision']:.4f}")
            print(f"  Recall:    {metrics['recall']:.4f}")
            print(f"  F1-score:  {metrics['f1-score']:.4f}")
            print('-' * 30)
    print('')
    print('')
    cm_1 = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm_1, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=np.unique(y),
            yticklabels=np.unique(y))
    plt.xlabel('Предсказанные значения', fontsize=11)
    plt.ylabel('Истинные значения', fontsize=11)
    plt.title(f'Матрица ошибок для модели {model_name}', fontsize=16)
    plt.show()
```

## `DecisionTreeClassifier`

В ходе исследовательского анализа было выяснено, что в данных присутствует небольшой дисбаланс классов, поэтому при обучении модели `DecisionTreeClassifier` мы используем параметр `class_weight='balanced'`. Это автоматически назначит веса классам обратно пропорционально их частоте. Также при кросс-валидации мы используем параметр `scoring='f1_weighted'`, так как нам нужна взвешенная оценка для подбора гиперпараметров. При расчёте метрик выбор параметра `average='weighted'` будет гарантировать, что метрики будут рассчитаны корректно, даже если классы несбалансированы

```python
# создадим словарь с гиперпараметрами

dt_params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': ['sqrt', 'log2', None],
    'class_weight': [None, 'balanced']
}

# инициализируем модель

dt = DecisionTreeClassifier(random_state=RANDOM_STATE, class_weight='balanced')
```

```python
# выполним подбор гиперпараметров и выведем их на экран

gs_1 = GridSearchCV(dt, dt_params, cv=5, scoring='f1_weighted')
gs_1.fit(X_train, y_train)

print('Лучшие параметры модели DecisionTreeClassifier:\n\n', gs_1.best_params_)
```

```python
# выполним предсказание модели на тренировочной и тестовой выборках

best_dt = gs_1.best_estimator_
y_train_pred_dt = best_dt.predict(X_train)
y_test_pred_dt = best_dt.predict(X_test)
```

```python
# выведем результаты обучения на экран

model_results('DecisionTreeClassifier', y_train, y_train_pred_dt, y_test, y_test_pred_dt)
```

Для нас предсказание высокого риска гораздо более важно, чем предсказание двух других классов — модель не должна допускать ситуаций, когда высокий риск при беременности принимается за незначительный. Метрика Recall говорит нам о том, что модель верно предсказала высокий риск в 91% случаев, когда он на самом деле был высоким

Инициализируем и обучим модель DummyClassifier, а затем сравним её метрику с метрикой нашей лучшей модели, чтобы оценить адекватность проделанной работы. DummyClassifier предсказывает средние значения, и если его метрики будут выше метрик нашей модели, это будет значить, что даже эта базовая модель работает лучше, а получившаяся модель CatBoostClassifier не принесёт пользы

```python
def dummy_classifier_test(best_model_name):
    
    """
    Функция dummy_classifier_test инициализирует модель DummyClassifier,
    обучает её на тренировочной выборке, выполняет предсказания на тестовой выборке,
    вычисляет Accuracy, Precision, Recall и F1, а затем строит сравнительную таблицу для тестовых значений
    лучшей модели и модели DummyClassifier
    
    Аргументы:
    best_model_name (str) : название модели
    
    Возвращает:
    dummy_result (DataFrame) : таблица со сравнительным результатом
    
    """
    
    dummy = DummyClassifier(strategy='most_frequent')
    dummy.fit(X_train, y_train)
    dummy_test_pred = dummy.predict(X_test)
    
    dummy_accuracy_test = accuracy_score(y_test, dummy_test_pred).round(4)
    dummy_precision_test = precision_score(y_test, dummy_test_pred, average='weighted', zero_division=0).round(4)
    dummy_recall_test = recall_score(y_test, dummy_test_pred, average='weighted', zero_division=0).round(4)
    dummy_f1_test = f1_score(y_test, dummy_test_pred, average='weighted', zero_division=0).round(4)
    
    cb_accuracy_test = accuracy_score(y_test, y_test_pred_cb).round(4)
    cb_precision_test = precision_score(y_test, y_test_pred_cb, average='weighted').round(4)
    cb_recall_test = recall_score(y_test, y_test_pred_cb, average='weighted').round(4)
    cb_f1_test = f1_score(y_test, y_test_pred_cb, average='weighted').round(4)
    
    dummy_result = {
        'model': [best_model_name, 'Dummy Classifier'],
        'Accuracy': [cb_accuracy_test, dummy_accuracy_test],
        'Precision': [cb_precision_test, dummy_precision_test],
        'Recall': [cb_recall_test, dummy_recall_test],
        'F1': [cb_f1_test, dummy_f1_test]
    }
    
    dummy_result = pd.DataFrame(dummy_result)
    
    return dummy_result
```

```python
# применим функцию dummy_classifier_test

dummy_result = dummy_classifier_test('Cat Boost Classifier')
```

```python
# оценим результат

print('Таблица-сравнение метрик у моделей CatBoostClassifier и DummyClassifier:')
dummy_result
```

Проверка прошла успешно: очевидно, наша лучшая модель не является бесполезной. Чем метрики выше, тем лучше — а у DummyClassifier они гораздо ниже
